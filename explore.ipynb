```python
# Files to load
student_data_to_load = "Conditions_Contributing_to_COVID-19_Deaths__by_State_and_Age__Provisional_2020-2022.csv"
# Read the file and store it in a Pandas DataFrame.
df = pd.read_csv(student_data_to_load)
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
df = df[df.State != "United States"]
df.head()
```

```python
df.rename(columns={'Age Group': 'ageGroup'}, inplace=True)
df.head()
```

```python
df = df[df.ageGroup != "Not stated"]
```

```python
df = df[df.ageGroup != "All Ages"]
```

```python
for col in df:
  print(df[col].unique())
```

```python
df.dtypes
```

```python
def change_string(age):
    if age == "0-24":
        return 1
    elif age == "25-34":
        return 2
    elif age == "35-44":
        return 3
    elif age == "45-54":
        return 4
    elif age == "55-64":
        return 5
    elif age == "65-74":
        return 5
    elif age == "75-84":
        return 6
    elif age == "85+":
        return 7
    else:
        return 8
df["ageGroup"] = df["ageGroup"].apply(change_string)
df.head()
```

```python
df = df.reset_index()
del df['index']
df.head()
```

```python
df.to_csv('cleaned.csv')
```
